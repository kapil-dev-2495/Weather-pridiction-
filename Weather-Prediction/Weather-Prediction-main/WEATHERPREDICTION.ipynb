```python
data['Pressure (millibars)'].hist()
```

```python
pressure_median = data['Pressure (millibars)'].mean()
      
def pressure(x):
    if x==0:
        return x + pressure_median
    else:
        return x
        
```

```python
data["Pressure (millibars)"] = data.apply(lambda row:pressure(row["Pressure (millibars)"]) , axis = 1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
y = data['Temperature (C)']
```

```python
X = data[['Humidity','Visibility (km)','Pressure (millibars)']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 101)
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
lm.coef_
```

```python
predictions = lm.predict(X_test)
```

```python
predictions
```

```python
plt.scatter(y_test, predictions)
plt.xlabel("y_test")
plt.ylabel("predictions")
```

```python
from sklearn import metrics
```

```python
print("MAE: ", metrics.mean_absolute_error (y_test, predictions))
print("MSE: ", metrics.mean_squared_error (y_test, predictions))
print("RMSE: ", np.sqrt(metrics.mean_squared_error (y_test, predictions)))
```

```python
sns.distplot((y_test-predictions), bins = 30)
```
